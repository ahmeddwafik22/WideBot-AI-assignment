```python
x_train, x_test, y_train, y_test = train_test_split(x, y,stratify=y, test_size=0.2, random_state=42)
```

# ML Models with TF-IDF Feature Extraction 

```python
def inference(model,x_test,y_test):
    y_pred=model.predict(x_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy}")
    
    # Calculate precision
    precision = precision_score(y_test, y_pred,average='macro')
    print(f"precision: {precision}")
    
    # Calculate recall
    recall = recall_score(y_test, y_pred,average='macro')
    print(f"recall: {recall}")
    
    # Calculate r
    f1 = f1_score(y_test, y_pred,average='macro')
    print(f"f1-score: {f1}")
    
    
    
    
    
```

```python
##TF-IDF assigns weights to words based on their importance in a document within a corpus
tfidf_vectorizer = TfidfVectorizer(
    max_df=0.60,         # Exclude terms that appear in more than 60% of the documents
    min_df=5,            # Exclude terms that appear in less than 5 documents
    max_features=35000,   # Limit the number of features to 35000
    ngram_range=(1, 2)   # Include unigrams and bigrams
)
```

```python
x_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
x_test_tfidf=tfidf_vectorizer.transform(x_test)
```

## Logisitc Regression with TF-IDF

```python
model = LogisticRegression(C=0.5) ## with regalization parameter c if it is too big "overfitting",too small "underfitting"
model.fit(x_train_tfidf, y_train)
```

```python
inference(model,x_test_tfidf,y_test)
```

## Random Forest with TF-IDF

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train_tfidf, y_train)
```

```python
inference(model,x_test_tfidf,y_test)
```
